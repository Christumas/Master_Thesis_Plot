```python
df2 = pd.concat( [df['a'], df['r'], df['giD']], axis = 1 ) 

def arrayconvstr(strarray):
        return eval(strarray)
df2['a'] = df2['a'].apply(arrayconvstr)
df2 ['r'] = df2['r'].apply(arrayconvstr)

df2 = df2.explode(['a','r'])
df2['Obs mean'] = df2['r']

def obsmean(group):
    group['Obs mean'] = group.groupby( (group['a'] != group['a'].shift()).cumsum(), group_keys=True )['r'].expanding().mean().reset_index(level=0 , drop = True)
    return group

df2 = df2.groupby('giD').apply(obsmean).reset_index(drop = True)
df2
```

```python
#functions 

def cbandit_trial(old_df,new_df):
    def arrayconvstr(strarray):
        return eval(strarray)
    new_df['Chosen Bandit per Trial'] = old_df['a'].apply(arrayconvstr)
    new_df['Rewards'] = old_df['r'].apply(arrayconvstr)
    return new_df.explode(['Chosen Bandit per Trial', 'Rewards'])

def moreinfo_bandit(new_df):
    new_df.loc[new_df['Chosen Bandit per Trial'] == 2, 'More Info Bandit'] = 1
    new_df.loc[new_df['Chosen Bandit per Trial'] == 1, 'More Info Bandit'] = 2
                       
                       
def get_game(old_df, new_df):
    #trials = old_df['giD']
    new_df['Game ID'] = old_df['giD']

def get_horizon(old_df,new_df):
    #horizons = old_df.loc[old_df['gamelength'].isin([5,10]), 'gamelength']
    horizons = old_df['gamelength']
    #print(horizons)
    new_df['Horizon'] = horizons
    #print(new_df['Horizon'])
    
    #changing 5 and 10 to short and long horizons
    new_df.loc[new_df['Horizon'] == 5,'Horizon'] = 'Short'
    new_df.loc[new_df['Horizon'] == 10,'Horizon'] = 'Long'

def get_means(old_df,new_df):
    def arrayconv(strarray):
        strarray = strarray.replace(':',',')
        listu = ast.literal_eval(strarray)
        array = np.array(listu, dtype = int)
        return array
    new_df['Reward Means'] = old_df['mean'].apply(arrayconv)
    
def mean_diff(new_df):
    difference = lambda array : (array[0]-array[1])
    new_df['Real Mean difference - Bandits'] = new_df['Reward Means'].apply(difference)
    
def chosen_bandit(old_df,new_df):
    #convert the string arrays to 
    def arrayconv(strarray):
        listu = ast.literal_eval(strarray)
        array = np.array(listu, dtype = int)
        return array[4] # we need the fifth element of the array #first free choice
    new_df['Chosen Bandit FFT'] = old_df['a'].apply(arrayconv)
    new_df.loc[new_df['Chosen Bandit FFT'] == 2, 'Chosen Bandit FFT'] = 'R' #2 is the right bandit
    new_df.loc[new_df['Chosen Bandit FFT'] == 1, 'Chosen Bandit FFT'] = 'L' #1 is the left bandit

def get_higherbandit(new_df):
    new_df['Higher Mean Bandit'] = new_df['Reward Means'].apply(np.argmax) #returns the index of the higher value from our array
    new_df.loc[new_df['Higher Mean Bandit'] == 1, 'Higher Mean Bandit'] = 'R'# 1 = Right bandit
    new_df.loc[new_df['Higher Mean Bandit'] == 0, 'Higher Mean Bandit'] = 'L'# 0 = Left bandit
    
def get_context(new_df):    
    new_df['Context'] = new_df['Reward Means'].apply(np.sum)
    new_df['Context'] = np.where(new_df['Context'] >0, 'Win', 'Loss')
    
    
def get_infocond(old_df,new_df):
    new_df['Information Condition'] = old_df['nforced'].apply(lambda x: np.array(ast.literal_eval(x), dtype = int))
    def check_cond(array):
        value, counts = np.unique(array,return_counts=True)
        #print(array,value,counts,type(counts))
        if counts[0] > counts[1] or counts[1] > counts[0]:
            return 'Unequal'
        elif counts[0] == counts[1]:
            return 'Equal'
            
    new_df['Information Condition'] = new_df['Information Condition'].apply(check_cond)
    

def gettrial(new_df):
    new_df['Trial Number'] = new_df.groupby('Game ID').cumcount() + 1



    
dataframe = cbandit_trial(df,dataframe)
get_game(df,dataframe)
get_horizon(df,dataframe)
get_means(df,dataframe)
mean_diff(dataframe)
chosen_bandit(df,dataframe)
get_higherbandit(dataframe)
get_context(dataframe)
get_infocond(df,dataframe)
gettrial(dataframe)
moreinfo_bandit(dataframe)



#need columns for all the predictors for your statistics
#trial is one instance of you picking a bandit
#game iD is the number of rounds (80 games), trial number should increase within each game depending on the horizon
#make a row for each action, and basically all the information you have about that action (informativeness, observed mean, 
#chosen bandit etc)
#make a column for informativeness
#for observed mean diff you need column r and a 
#make a column for obs mean for unchosen option and chosen option 




```
