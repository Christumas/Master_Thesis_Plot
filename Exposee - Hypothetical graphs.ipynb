```python
#hypothesis 1a = decision noise
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-30, 30)

# Define logistic functions with different steepness
def logistic_function(x, k, x0=0, c=0):
    return 1 / (1 + np.exp(-k * (x - x0))) + c

# Parameters for the logistic functions
k1 = 0.2 # Steep curve
k2 = 0.085# Less steep curve
k3 = 0.09 # Even less steep curve

c1 = 0.0
c2 = 0.0
c3 = 0.0


# Generate the logistic curves
y1 = logistic_function(x, k1, c = c1)
y2 = logistic_function(x, k2, c= c2)
y3 = logistic_function(x, k3, c = c3)

# Plotting the curves
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='NW', color='black')
plt.plot(x, y2, label='BED', color='gold')
plt.plot(x, y3, label='OB', color='blue')
plt.xlabel('Difference in means')
plt.ylabel('Probability of choosing left bandit')
plt.title('Decision Noise')
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.savefig('Hyp 1a',  bbox_inches='tight')
```

```python
#hypothesis 1b = random exploration
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-30, 30)

# Define logistic functions with different steepness
def logistic_function(x, k, x0=0, c=0):
    return 1 / (1 + np.exp(-k * (x - x0))) + c

# Parameters for the logistic functions
k1 = 0.14# Steep curve
k2 = 0.08# Less steep curve
k3 = 0.085 # Even less steep curve

k_h1 = [k1,k2,k3]
k_h6 = [k1-0.06,k2-0.03,k3-0.03]

c1 = 0.01
c2 = 0.04
c3 = 0.03

c_list = [c1,c2,c3]

curve_colors = ['black','gold', 'blue']
groups = ['NW','BED','OB']

fig, axes = plt.subplots(1,3, figsize = (12,3))

for i, ax_row in enumerate(axes.flatten()):
    print(i)
    ax_row.plot(x, logistic_function(x, k_h1[i], c= c_list[i]), color = curve_colors[i], label = 'Horizon 1')
    ax_row.plot(x, logistic_function(x, k_h6[i], c= c_list[i]), color = 'magenta', label = 'Horizon 6')
    ax_row.set_ylim(0,1.105)
    ax_row.margins(y = 0.1)
    ax_row.set_title(groups[i])
    ax_row.legend()




fig.text(0.5, 0.007, 'Observed differences in mean \n of the bandits(left - right)', ha='center', va='center')
fig.text(0.001, 0.5, 'Probability of choosing left bandit', ha='center', va='center', rotation='vertical')
plt.tight_layout()

plt.subplots_adjust(wspace=0.4)  
        
plt.savefig('Hyp 1b',  bbox_inches='tight')
```
